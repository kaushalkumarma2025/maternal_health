```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.stats.diagnostic import het_breuschpagan
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Load pivoted data

df = df.dropna(subset=['institutional_births', 'female_literacy', 'child_marriage',
                        'health_insurance', 'improved_sanitation', 'clean_fuel'])

print(f"Districts with complete data: {len(df)}")
print(df[['institutional_births', 'female_literacy', 'child_marriage',
           'health_insurance', 'improved_sanitation']].describe().round(2))

# ============================================================
# MODEL 1: Simple bivariate — literacy alone
# ============================================================
m1 = smf.ols('institutional_births ~ female_literacy', data=df).fit(cov_type='HC1')
print("\n=== MODEL 1: Bivariate ===")
print(m1.summary2().tables[1].round(4))
print(f"R-squared: {m1.rsquared:.4f}")

# ============================================================
# MODEL 2: Multiple controls, no state FE
# ============================================================
m2 = smf.ols('''institutional_births ~ female_literacy + child_marriage 
                + health_insurance + improved_sanitation + clean_fuel''', data=df).fit(cov_type='HC1')
print("\n=== MODEL 2: Multiple Controls ===")
print(m2.summary2().tables[1].round(4))
print(f"R-squared: {m2.rsquared:.4f}")

# ============================================================
# MODEL 3: State Fixed Effects
# ============================================================
m3 = smf.ols('''institutional_births ~ female_literacy + child_marriage 
                + health_insurance + improved_sanitation + clean_fuel
                + C(State)''', data=df).fit(cov_type='HC1')
print("\n=== MODEL 3: State Fixed Effects ===")
# Print only non-state-dummy coefficients
coefs = m3.summary2().tables[1]
non_state = coefs[~coefs.index.str.startswith('C(State)')]
print(non_state.round(4))
print(f"R-squared: {m3.rsquared:.4f}")

# ============================================================
# DIAGNOSTICS
# ============================================================
print("\n=== DIAGNOSTICS (Model 2) ===")

# VIF
X = df[['female_literacy', 'child_marriage', 'health_insurance', 
        'improved_sanitation', 'clean_fuel']].dropna()
X_const = sm.add_constant(X)
vif = pd.DataFrame({
    'Variable': X.columns,
    'VIF': [variance_inflation_factor(X_const.values, i+1) for i in range(len(X.columns))]
})
print("\nVIF (>10 = problem):")
print(vif.to_string(index=False))

# Breusch-Pagan heteroskedasticity test
bp_test = het_breuschpagan(m2.resid, m2.model.exog)
print(f"\nBreusch-Pagan p-value: {bp_test[1]:.4f}")
print(f"  {'Heteroskedasticity detected → robust SEs already applied' if bp_test[1] < 0.05 else 'Homoskedastic'}")

# RESET test (Ramsey)
from statsmodels.stats.diagnostic import linear_reset
reset = linear_reset(m2, power=3, use_f=True)
print(f"\nRESET test p-value: {reset.pvalue:.4f}")
print(f"  {'Possible specification error' if reset.pvalue < 0.05 else 'No specification error detected'}")

# ============================================================
# COMPARISON TABLE
# ============================================================
print("\n=== MODEL COMPARISON ===")
print(f"{'':30s} {'M1 (Bivariate)':>15s} {'M2 (Controls)':>15s} {'M3 (State FE)':>15s}")
print("-" * 78)
for var in ['female_literacy', 'child_marriage', 'health_insurance', 'improved_sanitation', 'clean_fuel']:
    vals = []
    for m in [m1, m2, m3]:
        if var in m.params:
            coef = m.params[var]
            pval = m.pvalues[var]
            stars = '***' if pval < 0.01 else '**' if pval < 0.05 else '*' if pval < 0.1 else ''
            vals.append(f"{coef:.3f}{stars}")
        else:
            vals.append("—")
    print(f"{var:30s} {vals[0]:>15s} {vals[1]:>15s} {vals[2]:>15s}")
print("-" * 78)
print(f"{'R-squared':30s} {m1.rsquared:>15.4f} {m2.rsquared:>15.4f} {m3.rsquared:>15.4f}")
print(f"{'N':30s} {int(m1.nobs):>15d} {int(m2.nobs):>15d} {int(m3.nobs):>15d}")
```

# Regression Results: Determinants of Institutional Delivery Across 707 Indian Districts

## Key Findings

### 1. Female Literacy is the Strongest and Most Robust Predictor

Female literacy is statistically significant at the 1% level across all three model specifications, with coefficients of 0.274 (bivariate), 0.145 (controls), and 0.330 (state FE). In the full state fixed-effects model, a 10 percentage-point increase in female literacy is associated with a 3.3 percentage-point increase in institutional delivery rates, holding all else constant. The coefficient dips in Model 2 (0.145) when correlated controls like clean fuel absorb shared variance, but rebounds to 0.330 once state fixed effects partial out between-state heterogeneity — revealing literacy's true within-state explanatory power.

**Policy implication:** Investments in girls' education yield downstream returns in maternal healthcare utilization. Districts with the lowest female literacy rates should be co-targeted for both education and maternal health interventions.

### 2. Clean Fuel Usage as a Proxy for Household Modernization

Clean fuel for cooking (LPG, electricity) emerges as the second-strongest predictor (coefficient: 0.161, p<0.01 in Model 3). This variable proxies for household wealth, connectivity to government distribution networks (e.g., Ujjwala Yojana), and broader modernization — all of which correlate with healthcare-seeking behavior. The coefficient attenuates from 0.277 (Model 2) to 0.161 (Model 3) upon adding state fixed effects, indicating that a substantial portion of the clean fuel–delivery association operates through state-level development differences.

**Policy implication:** Infrastructure development and welfare scheme penetration indirectly improve maternal health outcomes. Cross-sectoral convergence between energy access programs and maternal health programs could amplify impact.

### 3. Health Insurance Emerges as a Significant Predictor

Health insurance coverage is statistically significant in both Model 2 (0.121, p<0.01) and Model 3 (0.137, p<0.01). Unlike clean fuel, the coefficient actually increases after adding state fixed effects, suggesting that within-state variation in insurance coverage independently predicts institutional delivery. A 10 percentage-point increase in health insurance coverage is associated with a 1.4 percentage-point increase in institutional delivery rates. This likely captures the effect of schemes like Ayushman Bharat and state-specific health insurance programs that reduce financial barriers to facility-based delivery.

**Policy implication:** Expanding health insurance coverage — particularly in districts with low enrollment — can directly improve institutional delivery uptake by removing out-of-pocket cost barriers.

### 4. Child Marriage Effect is Statistically Null

Child marriage is statistically insignificant in both Model 2 (-0.008, p=0.83) and Model 3 (0.002, p=0.96). With the full 707-district dataset, the variable shows no independent association with institutional delivery in any specification. Child marriage and institutional delivery are correlated at the state level, but district-level child marriage rates have no independent predictive power.

**Policy implication:** Addressing child marriage requires state-level policy interventions (enforcement of PCMA, awareness campaigns) rather than district-level targeting within maternal health programs.

### 5. Sanitation's Negative Sign Attenuates and Loses Significance

Improved sanitation carries a negative coefficient (-0.057, p=0.10 in Model 3), no longer statistically significant at conventional levels. In the previous 340-district analysis (Phase 1 states only), this was significant (p<0.05) and posed an interpretation challenge. With the full dataset, the effect washes out — suggesting the earlier result was driven by sample composition. The residual negative direction may reflect ceiling effects in urbanized districts, but the lack of significance means no strong inference should be drawn.

### 6. State-Level Policy Variation Explains the Majority of Disparities

R-squared increases from 0.34 (Model 2) to 0.75 (Model 3) upon adding state fixed effects. This 41 percentage-point jump indicates that unobserved state-level factors are the dominant driver of institutional delivery variation — including differential implementation of Janani Suraksha Yojana (JSY), National Health Mission (NHM) funding allocations, state health budgets, ASHA worker density, and political prioritization of maternal health.

**Policy implication:** The state a woman resides in is the single largest determinant of whether she delivers in an institution. Reducing interstate disparities in maternal health program implementation should be a central focus of national policy. States like Nagaland, Meghalaya, and Arunachal Pradesh require disproportionate attention and resource allocation.

## Limitations

This analysis uses district-level aggregate data from NFHS-5 factsheets covering all 707 districts across 36 states and union territories. Results represent ecological associations, not individual-level causal effects. The RESET test indicates potential specification issues, suggesting that the relationship between predictors and institutional delivery may be non-linear or that relevant interaction terms are omitted. The constraints warning in Model 3 reflects automatic dropping of one state as the reference category — standard fixed-effects behavior that does not affect coefficient estimates. Individual-level analysis using DHS microdata (pending access approval) will address ecological inference limitations through logistic regression with household-level controls and survey weights.
